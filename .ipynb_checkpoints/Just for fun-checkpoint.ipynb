```python
num = int(input())
if num % 3 == 0 and num % 5 == 0:
    print(num,': ',3,sep='')
elif num % 5 == 0:
    print(num,': ',2,sep='')
elif num % 3 == 0:
    print(num,': ',1,sep='')
else:
    print(num,': ',0,sep='')
```

```python
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.optimizers import Adam
from keras.callbacks import ReduceLROnPlateau
```

```python
x = np.zeros((1000,10))
```

```python
x[1,9]
```

```python
def create_data(sample_num):
    if sample_num > 1000:
        print('out of number')
        return 
    x = np.zeros((sample_num,10))
    y = np.zeros((sample_num,4))
    for num in range(1,sample_num+1):
        temp = num
        for digit_x in range(9,-1,-1):
            x[num-1,digit_x] = temp // (2**digit_x)
            temp = temp % (2**digit_x)
        if num % 3 == 0 and num % 5 == 0:
            y[num-1,3] = 1
        elif num % 5 == 0:
            y[num-1,2] = 1
        elif num % 3 == 0:
            y[num-1,1] = 1
        else:
            y[num-1,0] = 1
    return x,y
```

```python
x,y = create_data(1000)
```

```python
X_train = x[101:,:]
Y_train = y[101:,:]
X_test = x[:101,:]
Y_test = y[:101,:]
```

```python
model = Sequential()

model.add(Dense(input_dim=10, units=1000))
model.add(Activation('relu'))
model.add(Dense(units=4, activation = "softmax"))

optimizer = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08)
model.compile(optimizer = optimizer , loss = "categorical_crossentropy", metrics=["accuracy"])

learning_rate_reduction = ReduceLROnPlateau(monitor='val_acc', 
                                            patience=3, 
                                            verbose=1, 
                                            factor=0.5, 
                                            min_lr=0.000001)
```

```python
model.fit(X_train,Y_train,batch_size=20,epochs=300,callbacks=[learning_rate_reduction])
test_result = model.evaluate(X_test,Y_test,batch_size=1000)
print('\nTest acc:',test_result[1])
```
